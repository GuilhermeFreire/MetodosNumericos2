```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A = np.array([[3,2],[2,6]])
A = np.array([[3,0],[0,-6]])
b = np.array([[2],[-8]])

f = lambda x: 0.5 * x.T.dot(A.dot(x)) - b.T.dot(x)
grad_f = lambda x: A.dot(x) - b

np.random.seed(2)
x = np.random.randn(2,1)
# x = np.array([[3], [3.5]])
hist = []
for i in range(40):
    hist.append(x)
    g_k = grad_f(x)
    alpha = (g_k.T.dot(g_k)) / (g_k.T.dot(A.dot(g_k)) + 1e-8)
    x = x - alpha*g_k
    print(i, x, end="\n\n")
    
hist = np.array(hist)
```

```python
E1 = np.linspace(-8,8,100)
E2 = np.linspace(-8,8,100)
E1, E2, = np.meshgrid(E1,E2)
Z = np.zeros_like(E1)
for i in range(E1.shape[0]):
    for j in range(E1.shape[1]):
        x = np.array([E1[i][j], E2[i][j]])
        Z[i][j] = f(x)
        
plt.figure(figsize=(10,10))
levels = np.linspace(-50,40,30)
plt.contour(E1, E2, Z)
plt.scatter(hist[:,0], hist[:,1], c="red")
plt.plot(hist[:,0], hist[:,1], c="red")
plt.show()
```

```python
np.random.seed(1234)
size = 1000
A = np.random.rand(size,size)*2 - 1
A = A.T.dot(A)
A = np.diag(np.arange(size))
b = np.random.rand(size,1)

f = lambda x: 0.5 * x.T.dot(A.dot(x)) - b.T.dot(x)
grad_f = lambda x: A.dot(x) - b

np.random.seed(2)
x = np.random.randn(size,1)
# x = np.array([[3], [3.5]])
hist = []
erros = []
for i in range(200):
    hist.append(x)
    g_k = grad_f(x)
    alpha = (g_k.T.dot(g_k)) / (g_k.T.dot(A.dot(g_k)) + 1e-8)
    x = x - alpha*g_k
    erros.append(np.linalg.norm(x - hist[-1]) / np.linalg.norm(hist[-1]))
#     print(i, np.linalg.norm(x - hist[-1]), end="\n\n")
    
hist = np.array(hist)
plt.figure(figsize=(10,10))
plt.semilogy(erros)
plt.show()
```
