## Exercício 1

Pseudo inversa

Primeiro, montemos um dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
strengh = 0.1
low_b, high_b = 0, 1.2
x = np.linspace(low_b, high_b, 10)
y = 2 * x**3 - 3 * x**2 + x + (np.random.rand(x.shape[0]) - 0.5) * strengh

plt.scatter(x, y)
plt.show()
```

Agora, implementemos o método da pseudo-inversa.

```python
def build_Vandermonde(x, degree):
    A = np.ones((x.shape[0], degree+1))
    for d in range(1, degree+1):
        A[:,d] = x**d
    return A
```

```python
def calc_pol(x, coef):
    res = np.zeros_like(x)
    for i, c in enumerate(coef):
        res += c * x**i
    return res
```

```python
dim = 3
A = build_Vandermonde(x, dim)
U, S, Vt = np.linalg.svd(A)

S_inv = np.zeros((Vt.shape[0], U.shape[0]))
non_null_vals = S.shape[0]
S_inv[:non_null_vals, :non_null_vals] = np.diag(1 / S)

pseudo_inv_A = Vt.T.dot((S_inv.dot(U.T)))
sol = pseudo_inv_A.dot(y)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

Comparando o resultado com o do Mínimos Quadrados (um pouco abaixo na página), vemos que são iguais. Isso ocorreu quando tentamos um polinômio de grau igual ao de geração dos pontos. Tentemos agora um de grau maior.

```python
dim = 10
A = build_Vandermonde(x, dim)
U, S, Vt = np.linalg.svd(A)

S_inv = np.zeros((Vt.shape[0], U.shape[0]))
non_null_vals = S.shape[0]
S_inv[:non_null_vals, :non_null_vals] = np.diag(1 / S)

pseudo_inv_A = Vt.T.dot((S_inv.dot(U.T)))
sol = pseudo_inv_A.dot(y)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

Novamente a solução encontrada foi a mesma da dos Mínimos Quadrados. Isso se deve ao fato, de que esse método realmente encontra a inversa quando a matriz é inversível. Testemos então com pontos que tornem essa matriz não inversível.

```python
np.random.seed(1234)
strengh = 0.1
low_b, high_b = 0, 1.2
x2 = np.concatenate((np.linspace(low_b, high_b, 10), np.array([0.4, 0.8, 1.2])))
y2 = 2 * x2**3 - 3 * x2**2 + x2 + (np.random.rand(x2.shape[0]) - 0.5) * strengh


plt.scatter(x2, y2)
plt.show()
```

Rodemos agora, o método da pseudo inversa.

```python
dim = 3
A = build_Vandermonde(x2, dim)
U, S, Vt = np.linalg.svd(A)

S_inv = np.zeros((Vt.shape[0], U.shape[0]))
non_null_vals = S.shape[0]
S_inv[:non_null_vals, :non_null_vals] = np.diag(1 / S)

pseudo_inv_A = Vt.T.dot((S_inv.dot(U.T)))
sol = pseudo_inv_A.dot(y2)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x2, y2)
plt.show()
```

Vemos que o método ainda se comporta bem. Entretanto o que acontece se elevarmos o grau do polinomio de aproximação?

```python
dim = 6
A = build_Vandermonde(x2, dim)
U, S, Vt = np.linalg.svd(A)

S_inv = np.zeros((Vt.shape[0], U.shape[0]))
non_null_vals = S.shape[0]
S_inv[:non_null_vals, :non_null_vals] = np.diag(1 / S)

pseudo_inv_A = Vt.T.dot((S_inv.dot(U.T)))
sol = pseudo_inv_A.dot(y2)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x2, y2)
plt.show()
```

Até o grau 6 preticamente não há diferença perceptível.

```python
dim = 7
A = build_Vandermonde(x2, dim)
U, S, Vt = np.linalg.svd(A)

S_inv = np.zeros((Vt.shape[0], U.shape[0]))
non_null_vals = S.shape[0]
S_inv[:non_null_vals, :non_null_vals] = np.diag(1 / S)

pseudo_inv_A = Vt.T.dot((S_inv.dot(U.T)))
sol = pseudo_inv_A.dot(y2)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x2, y2)
plt.show()
```

Agora com grau 7, notamos uma mudança de comportamento na função.

```python
dim = 9
A = build_Vandermonde(x2, dim)
U, S, Vt = np.linalg.svd(A)

S_inv = np.zeros((Vt.shape[0], U.shape[0]))
non_null_vals = S.shape[0]
S_inv[:non_null_vals, :non_null_vals] = np.diag(1 / S)

pseudo_inv_A = Vt.T.dot((S_inv.dot(U.T)))
sol = pseudo_inv_A.dot(y2)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x2, y2)
plt.show()
```

Em grau 9 já vemos um grande overfitting em nossos dados.

```python
dim = 10
A = build_Vandermonde(x2, dim)
U, S, Vt = np.linalg.svd(A)

S_inv = np.zeros((Vt.shape[0], U.shape[0]))
non_null_vals = S.shape[0]
S_inv[:non_null_vals, :non_null_vals] = np.diag(1 / S)

pseudo_inv_A = Vt.T.dot((S_inv.dot(U.T)))
sol = pseudo_inv_A.dot(y2)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x2, y2)
plt.show()
```

Finalmente, em grau 10, o método não resiste e a solução está sofrendo gravemente devido ao overfitting e não consegue mais uma representação decente da função original.

Para tentar mitigar esse problema, exploremo outra solução: a regularização de Tikhonov.

## Exercício 2

Regularização de Tikhonov

Primeiro montemos um dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
strengh = 0.1
low_b, high_b = 0, 1.2
x = np.linspace(low_b, high_b, 10)
y = 2 * x**3 - 3 * x**2 + x + (np.random.rand(x.shape[0]) - 0.5) * strengh

plt.scatter(x, y)
plt.show()
```

Agora, utilizando os conhecimentos da aula 11, podemos encontrar um polinômio que se adequa aos pontos usando a Regularização de Tikhonov.

```python
dim = 3
A = build_Vandermonde(x, dim)
alpha = 0.0
A_prime = A.T.dot(A) + alpha * np.identity(dim+1)
b_prime = A.T.dot(y)
sol = np.linalg.solve(A_prime, b_prime)
print(sol)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

Ao colocar o peso de $\alpha$ como $0$, obtemos a solução de Mínimos Quadrados, como era de se esperar. Entretanto, se vamos aumentando o peso da regularização, notamos uma característica interessante.

```python
dim = 3
A = build_Vandermonde(x, dim)
alpha = 0.001
A_prime = A.T.dot(A) + alpha * np.identity(dim+1)
b_prime = A.T.dot(y)
sol = np.linalg.solve(A_prime, b_prime)
print(sol)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

```python
dim = 3
A = build_Vandermonde(x, dim)
alpha = 0.01
A_prime = A.T.dot(A) + alpha * np.identity(dim+1)
b_prime = A.T.dot(y)
sol = np.linalg.solve(A_prime, b_prime)
print(sol)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

Notamos que conforme o peso vai aumentando, a curva vai ficando cada vez menos precisa e lembra apenas o jeitão da curva original. Se continuarmos deveríamos ver a curva se assemelhar cada vez mais de uma parábola e posteriormente de uma reta.

```python
dim = 3
A = build_Vandermonde(x, dim)
alpha = 0.1
A_prime = A.T.dot(A) + alpha * np.identity(dim+1)
b_prime = A.T.dot(y)
sol = np.linalg.solve(A_prime, b_prime)
print(sol)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

```python
dim = 3
A = build_Vandermonde(x, dim)
alpha = 60
A_prime = A.T.dot(A) + alpha * np.identity(dim+1)
b_prime = A.T.dot(y)
sol = np.linalg.solve(A_prime, b_prime)
print(sol)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

Esses testes foram realizados com um polinômio do mesmo grau do que o gerador dos pontos. Testemos agora com um polinômio aproximador de grau maior.

```python
dim = 10
A = build_Vandermonde(x, dim)
alpha = 0.0
A_prime = A.T.dot(A) + alpha * np.identity(dim+1)
b_prime = A.T.dot(y)
sol = np.linalg.solve(A_prime, b_prime)
print(sol)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

Usando apenas o Mínimos Quadrados, vemos que ocorre um overfit no conjunto de pontos. Para mitigar, vejamos o que acontece ao aumentar o valor da penalidade da regularização.

```python
dim = 10
A = build_Vandermonde(x, dim)
alpha = 0.005
A_prime = A.T.dot(A) + alpha * np.identity(dim+1)
b_prime = A.T.dot(y)
sol = np.linalg.solve(A_prime, b_prime)
print(sol)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x, y)
plt.show()
```

Ao aumentar apenas um pouco, conseguimos um resultado tão bom quanto o método de Mínimos Quadrados com o grau certo. Com isso, temos uma pequena prova do poder que uma regularização pode ter em um modelo de aproximação. Somos capazes de criar modelos extremamente genéricos (nesse caso um de grau 10 para representar dados oriundos de um de grau 3) sem nos preocuparmos com overfitting.

Testemos esse método, agora, no segundo dataset, no qual a matriz não é inversível.

```python
dim = 10
A = build_Vandermonde(x2, dim)
alpha = 0.005
A_prime = A.T.dot(A) + alpha * np.identity(dim+1)
b_prime = A.T.dot(y2)
sol = np.linalg.solve(A_prime, b_prime)
print(sol)

fit_X = np.linspace(low_b, high_b)
fit_Y = calc_pol(fit_X, sol)

plt.plot(fit_X, fit_Y)
plt.scatter(x2, y2)
plt.show()
```

Note que diferentemente do método da pseud-inversa, o método da regularização não tem problema nenhum com graus muito maiores do que o que originou o dado.
